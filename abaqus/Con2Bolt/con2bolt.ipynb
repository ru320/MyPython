```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import altair as alt
from numpy import pi
```

```python
 data = pd.read_csv('data.csv',delimiter=',')
```

```python
Time = data['Time']
temp_1 = data['Temp_1']
temp_2 = data['Temp_2']
temp_3 = data['Temp_3']
temp_4 = data['Temp_4']
temp_5 = data['Temp_5']
temp_6 = data['Temp_6']
```

```python
D = 20.
A = D**2*pi/4.
WB = pi*D**3/32.
RP = 640.
print(D,A,WB,RP)
```

```python
def srss(a,b):
    c = np.sqrt(a**2 + b**2)
    return c
def sigma_M(Sz,Sb,T):
    Sm = np.sqrt((Sz + Sb)**2 + 3 * (T**2))
    return Sm
def sigma_Z(Fax,A):
    Sz = Fax / A
    return Sz
def sigma_B(MB,WB):
    Sb = MB/WB
    return Sb
def Tau(FQ,A):
    T = FQ / A
    return T
```

```python
FQ = srss(temp_2, temp_3)
MB = srss(temp_5, temp_6)
```

```python
Sz = sigma_Z(temp_1, A )
Sb = sigma_Z(MB, WB )
T = sigma_Z(FQ, A )
Sm = sigma_M(Sz,Sb,T)
S = RP / Sm
U = 100 / RP * Sm

```

```python
plt.plot(Time, Sm, label='Smises')
plt.plot(Time, Sz, label='Szug')
plt.plot(Time, Sb, label='Sbiegung')
plt.plot(Time, T, label='Tau')
plt.legend()
plt.show()
```

```python
plt.plot(Time, S, label='S')
plt.plot(Time, U, label='A')
plt.legend()
plt.show()
```
